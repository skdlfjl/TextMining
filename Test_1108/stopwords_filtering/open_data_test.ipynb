```python
## 우선 tokens6 데이터로 테스트 해본다

data = open('tokens6.txt', 'r', encoding='utf-8')
data = data.readlines()
print(len(data))

for i in range(len(data)):
    data[i] = data[i].replace('\n', '')

print(data)   ## 잘 불러와지네요
```

```python
# token등장 빈도수 그래프 그려보기
import nltk

tokens6 = data
a = nltk.FreqDist(tokens6)
a.plot(10000)

## 너무 많이(적게) 등장한 token 제거가 필요하다! (Week3_Lec3_전처리 참고)
```
