```python
# 파이썬개발자[6] 데이터엔지니어[7] 데이터사이언티스트[12] 머신러닝엔지니어[13] 빅데이터엔지니어[15]
category = [6, 7, 12, 13, 15]

# 각 카테고리의 데이터 개수(채용공고의 개수)를 가져와서 튜플형태로 저장한 뒤, 리스트에 넣는다
category_len = []
for category_num in category:
    data = open('../category{0}.txt'.format(category_num), 'r', encoding = 'utf-8')
    lines = data.readlines()  # 데이터 불러와서 line 단위로 넣는다
    #print(len(lines))
    tu = category_num, len(lines)
    category_len.append(tu)
    
print(category_len)# ( 카테고리번호, 크롤링 데이터 개수 ) 모양의 튜플

for category_num, data_num in category_len:
    
    data = open('../data_cleansing_final/cleansing_category{0}_list{1}.txt'.format(category_num, data_num), 'r', encoding = 'utf-8')
    data = data.readlines()  # 데이터 불러와서 line 단위로 넣는다
    
    for i in range(len(data)):
        data[i] = data[i].replace('\n', '')    # 각 문장마다 들어있는 줄바꿈 기호 없애주기
    
    data = ' '.join(data)
    
    ## 각 변수에 저장하자
    globals()['category{}'.format(category_num)] = data
```

```python
import json

class vocab(dict):
    _UNK_ = '<UNK>'
    _PAD_ = '<PAD>'
    def __init__(self):
        self['w2i'] = dict()
        self['i2w'] = dict()

    def add_unk(self):
        self.add(vocab._UNK_)
    def add_pad(self):
        self.add(vocab._PAD_)
        
    def exist(self, w):
        return (w in self['w2i'].keys())
    
    def index(self, w):
        return self['w2i'].get(w)
    def index_unk(self):
        return self.index(vocab._UNK_)
    def index_pad(self):
        return self.index(vocab._PAD_)
    
    def word(self, i):
        return self['i2w'].get(i)
    
    def add(self, w):
        if self.exist(w):
            return False
        new_index = len(self['i2w'])
        self['w2i'][w] = new_index
        self['i2w'][new_index] = w
        return True
    
    def size(self):
        return len(self['w2i'])
    
    def save(self, path):
        with open(path, 'r') as f:
            loaded = json.loads(f.read())
        self['w2i'] = loaded.get('w2i')
        self['i2w'] = loaded.get('i2w')
```
