```python
# hannanum을 사용할것이므로 채용공고별로 나눠진 데이터를 사용한다
data = open('../data_cleansing_final/cleansing_category6_list243.txt', 'r', encoding='utf-8')
data = data.readlines()  # 데이터 불러와서 저장한다
print('#크롤링한 채용공고의 개수 = ', len(data), '\n')

for i in range(len(data)):
    data[i] = data[i].replace('\n', '')

#print(data)  # 채용공고별로 나눠서 저장되어있는 데이터


# 단어 토큰화 - pos태그를 사용하지 않습니다
# 문서에서 단어의 순서가 중요하지 않으므로 문장 토큰화를 사용하지 않습니다.
from konlpy.tag import Hannanum
hannanum = Hannanum()

# 243번째 채용공고 토큰 확인해보기
print('#243번째 채용공고 hannanum 형태소 분석(추출) = ', len(hannanum.morphs(data[242])), '\n', hannanum.morphs(data[242]))
```

```python
token_list = []
for i in data:
    token_list.append(hannanum.morphs(i))
    
#print(len(token_list), token_list)  
# 리스트 안에 크롤링한 채용공고 개수만큼의 토큰 리스트들이 들어있음
```

```python
tokens = sum(token_list, [])  # 이중리스트 해제
print(tokens) # category6의 전체 토큰입니다
```

```python
# pos태깅을 이용한 stopwords 제거
# 주요 단어를 제외한 나머지 단어들을 제거해준다.
```

```python
token_pos_list = []
for i in data:
    token_pos_list.append(hannanum.pos(i))
    
#print(len(token_pos_list), token_pos_list)  
# 리스트 안에 크롤링한 채용공고 개수만큼의 토큰 리스트들이 들어있음

tokens_pos = sum(token_pos_list, [])  # 이중리스트 해제
print(tokens_pos[:20])
```

```python
tagset_list = []
for i in tokens_pos:
    tagset_list.append(i[1])  # tag들만 추출해서 리스트로 저장한다

tag = set(tagset_list)
print('#사용된 hannanum tagset :', tag)
# {'E', 'X', 'M', 'F', 'I', 'N', 'S', 'P', 'J'}
```

```python
#dir(hannanum)
type(hannanum.tagset)  # dict타입이므로 키를 이용하여 검색이 가능합니다. 

#hannanum.tagset

tag_dic = []
for i in tag:
    #print(i,':', hannanum.tagset[i])
    a = [i, hannanum.tagset[i]]
    tag_dic.append(a)

print(tag_dic)
```

```python
k_tokens = []
pos_stopwords = []
for i in range(len(tokens_pos)):
    if tokens_pos[i][1] == 'N' or tokens_pos[i][1] == 'F':
        k_tokens.append(tokens[i])
    else:
        pos_stopwords.append(tokens[i])
        
print(len(tokens), '->', len(k_tokens), '\n')
print('#전체 tokens :', tokens[:50], '\n')        
print('#postag를 이용해 불용어를 삭제한 tokens :', k_tokens[:50])
```

```python
# 한국어 stopwords 리스트 불러와서 제거
```

```python
stopwords_ko = open('stopwords_ko.txt', 'r', encoding='utf-8')
stopwords_ko = stopwords_ko.readlines()  # 데이터 불러와서 저장한다

for i in range(len(stopwords_ko)):
    stopwords_ko[i] = stopwords_ko[i].replace('\n', '')
    
#print(stopwords_ko)
```

```python
kk_tokens = [w for w in k_tokens if w not in stopwords_ko]
print('#한국어 stopwords 리스트 제거 :', len(k_tokens), '->', len(kk_tokens))
```

```python
# 영어 stopwords 제거
```

```python
# 불용어 목록 얻음 (소문자화 된 점에 유의)
import nltk
#nltk.download('stopwords')
stopwords_en = nltk.corpus.stopwords.words('english')
#print(len(stopwords_en), stopwords_en, '\n')

k_e_tokens = [w for w in kk_tokens if w.lower() not in stopwords_en]
# lower()를 이용해서 임시로 소문자화 한 뒤 필터링 한 단어(토큰)들을 k_e_tokens변수에 저장

print('#nltk 영어 stopwords 필터링 후 = ', len(kk_tokens), '->', len(k_e_tokens))  # 134개 토큰 삭제
```

```python
# 1글자 단어 제거(영어는 제외한다) - 1글자 stopwords 리스트 생성
# 저장하지 않고 각 카테고리에서 그때그때 만들어 쓴다
```

```python
# 1글자 토큰을 one_word 리스트에 저장한다
one_word = [] 
for i in k_e_tokens:
    if len(i)==1:
        #print(i)
        one_word.append(i)

u = set(one_word)  # 유니크 워드만 뽑기
```

```python
stopwords_one = []
for i in u:
    if ord('a') <= ord(i.lower()) <= ord('z'):
        print(i)
    else:
        stopwords_one.append(i)
        
print(stopwords_one) # 영어를 제외한 1글자 stopwords 리스트 stopwords_one
```

```python
final_tokens = [w for w in k_e_tokens if w not in stopwords_one]
# 필터링 한 단어(토큰)들을 final_tokens변수에 저장

print('#1글자 토큰(한국어) 필터링 후 = ', len(k_e_tokens), '->', len(final_tokens))
```

```python
# 전부 소문자로 바꿔주자
```

```python
low_tokens = []
for i in range(len(final_tokens)):
    low_tokens.append(final_tokens[i].lower())
```

```python
fdistPlain = nltk.FreqDist(final_tokens)      # stopwords 필터링
fdistPlain_low = nltk.FreqDist(low_tokens)    # 소문자로 바꿔준거

#print(fdistPlain.most_common(30), '\n')
print(fdistPlain_low.most_common(30))  # 빈도수가 높은 토큰 30개 추출
```
